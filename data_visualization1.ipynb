```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([0,10])
ypoints = np.array([20,100])
print(plt.plot(xpoints,ypoints))
print(plt.show())
```

```python
x = np.array([0,5])
y = np.array([0,25])
print(plt.plot(x,y,marker='D'))
print(plt.show())
```
